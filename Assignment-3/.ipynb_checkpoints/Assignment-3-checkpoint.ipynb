```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
path = r"student-mat.csv"
data = pd.read_csv(path, delimiter=';')
correlations = data.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlations, cmap='coolwarm', annot=False)
plt.show()
```

```python
from matplotlib import pyplot as plt
from pandas import read_csv
from pandas.plotting import scatter_matrix
path = r"student-mat.csv"
names = ['school', 'sex', 'age', 'address', 'famsize', 'Pstatus', 'Medu', 'Fedu', 'Mjob', 'Fjob', 'reason',
         'guardian', 'traveltime', 'studytime', 'failures', 'schoolsup', 'famsup', 'paid', 'activities', 'nursery',
         'higher', 'internet', 'romantic', 'famrel', 'freetime', 'goout', 'Dalc', 'Walc', 'health', 'absences', 'G1', 'G2', 'G3']
data = read_csv(path, delimiter=';', usecols=['age', 'Medu', 'Fedu', 'traveltime', 'studytime', 'failures', 'famrel', 'freetime', 'goout', 'Dalc', 'Walc', 'health', 'absences', 'G1', 'G2', 'G3'])
scatter_matrix(data, figsize=(10, 10))
plt.show()
```
